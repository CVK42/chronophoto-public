```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

# Chemin vers le dossier contenant les images
dossier_images = r'C:\Users\chris\OneDrive\Documents\Python Scripts\chronophoto\photos'

# Listes pour stocker les moments colorim√©triques et la moyenne de H
mean_H_values = []
annees = []

# Mesurer le temps de d√©but
temps_debut = time.time()

# Parcourir les fichiers dans le dossier
for fichier in os.listdir(dossier_images):
    if fichier.lower().endswith(('.jpg', '.jpeg')):
        chemin_image = os.path.join(dossier_images, fichier)

        # Extraire l'ann√©e du nom du fichier
        annee = int(fichier.split('-')[0])

        # Charger l'image en RGB
        img = cv2.imread(chemin_image)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Convertir l'image en HSV
        img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

        # Extraire la composante H
        H = img_hsv[:, :, 0]

        # Calculer la moyenne de la composante H
        mean_H = np.mean(H)
        mean_H_values.append(mean_H)
        annees.append(annee)

# Calculer la moyenne et la m√©diane de mean_H_values sur chaque ann√©e
moyenne_annuelle_H = []
mediane_annuelle_H = []
annees_uniques = np.unique(annees)

for annee_unique in annees_uniques:
    valeurs_annee = [mean_H_values[i] for i, annee in enumerate(annees) if annee == annee_unique]
    moyenne_annuelle_H.append(np.mean(valeurs_annee))
    mediane_annuelle_H.append(np.median(valeurs_annee))

# Tracer les graphiques de la moyenne et de la m√©diane c√¥te √† c√¥te
plt.figure(figsize=(10, 5))

# Graphique de la moyenne
plt.subplot(1, 2, 1)
plt.title('Moyenne de la composante H en fonction de l\'ann√©e')
plt.xlabel('Ann√©e')
plt.ylabel('Moyenne de la composante H')
plt.scatter(annees_uniques, moyenne_annuelle_H, marker='+', c='red')

# Graphique de la m√©diane
plt.subplot(1, 2, 2)
plt.title('M√©diane de la composante H en fonction de l\'ann√©e')
plt.xlabel('Ann√©e')
plt.ylabel('M√©diane de la composante H')
plt.scatter(annees_uniques, mediane_annuelle_H, marker='+', c='blue')

plt.tight_layout()
plt.show()

# Mesurer le temps de fin
temps_fin = time.time()
temps_execution = temps_fin - temps_debut
print("Le programme a mis", temps_execution, "secondes √† s'ex√©cuter.")
```
